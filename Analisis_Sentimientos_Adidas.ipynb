```python
from transformers import pipeline
```

```python
# Análisis de sentimientos
sentiment_analyzer = pipeline('sentiment-analysis')
```

```python
# Texto de entrada
text_sentiment = """
El objetivo principal de este proyecto en Power BI es analizar y visualizar las principales
tendencias de ventas de Adidas durante los años 2020 y 2021. A través de la creación de
dashboards interactivos y gráficos intuitivos, se busca proporcionar una comprensión clara
y detallada de la dinámica de ventas, identificando patrones, fluctuaciones estacionales y
puntos destacados en la actividad comercial.
"""
```

```python
# Dividir el texto en oraciones o párrafos
sentiment_segments = text_sentiment.split('.')
results_sentiment = [sentiment_analyzer(segment.strip()) for segment in sentiment_segments if segment.strip()]
print("Análisis de Sentimientos:")
for segment, result in zip(sentiment_segments, results_sentiment):
    if segment.strip():
        print(f"Texto: {segment.strip()}\nSentimiento: {result}\n")
```

```python
# Traducción de texto (de español a inglés)
translator = pipeline('translation', model='Helsinki-NLP/opus-mt-es-en')
text_translation = """
El objetivo principal de este proyecto en Power BI es analizar y visualizar las principales
tendencias de ventas de Adidas durante los años 2020 y 2021. A través de la creación de
dashboards interactivos y gráficos intuitivos, se busca proporcionar una comprensión clara
y detallada de la dinámica de ventas, identificando patrones, fluctuaciones estacionales y
puntos destacados en la actividad comercial.
"""
result_translation = translator(text_translation)
print("\nTraducción de Texto:")
print(result_translation)
```

```python
# Resumen de texto
summarizer = pipeline('summarization')
text_summary = """
El objetivo principal de este proyecto en Power BI es analizar y visualizar las principales
tendencias de ventas de Adidas durante los años 2020 y 2021. A través de la creación de
dashboards interactivos y gráficos intuitivos, se busca proporcionar una comprensión clara
y detallada de la dinámica de ventas, identificando patrones, fluctuaciones estacionales y
puntos destacados en la actividad comercial.
"""
result_summary = summarizer(text_summary, max_length=50, min_length=25, do_sample=False)
print("\nResumen de Texto:")
print(result_summary)
```

Otras alternativas a evaluar

Reconocimiento de entidades nombradas (NER)

```python
# Crear un pipeline para NER (Reconocimiento de Entidades Nombradas)
ner_pipeline = pipeline("ner", framework="tf", model="dbmdz/bert-large-cased-finetuned-conll03-english")

# Texto de entrada
text = """
El objetivo principal de este proyecto en Power BI es analizar y visualizar las principales tendencias de ventas de Adidas durante los años 2020 y 2021. A través de la creación de dashboards interactivos y gráficos intuitivos, se busca proporcionar una comprensión clara y detallada de la dinámica de ventas, identificando patrones, fluctuaciones estacionales y puntos destacados en la actividad comercial.
"""

# Realizar el reconocimiento de entidades nombradas
ner_results = ner_pipeline(text)

# Imprimir los resultados
print("Entidades Nombradas encontradas en el texto:")
for entity in ner_results:
    print(f"Entidad: {entity['word']}, Etiqueta: {entity['entity']}")
```

**Explicación de la salida**

Entidad: Power, Etiqueta: I-ORG: La entidad "Power" ha sido identificada como parte de una organización (ORG). La etiqueta "I-ORG" indica que esta es una entidad de organización y que es una parte continua de una entidad más larga.

Entidad: B, Etiqueta: I-ORG: La entidad "B" también ha sido identificada como parte de una organización (ORG). Al igual que la anterior, la etiqueta "I-ORG" indica que esta es una entidad de organización y que es una parte continua de una entidad más larga.

Entidad: ##I, Etiqueta: I-ORG: Esta entidad parece ser parte de una organización, pero la forma en que está tokenizada indica que el modelo considera que "I" es una parte de la entidad y "##I" puede ser un prefijo o una parte de una palabra que sigue a "I".

Entidad: Ad, Etiqueta: I-ORG: Similar a las anteriores, "Ad" ha sido identificado como parte de una organización (ORG).

Entidad: ##idas, Etiqueta: I-ORG: Al igual que "##I", "##idas" parece ser una parte de una palabra o un sufijo que sigue a una entidad de organización.

En resumen, el modelo ha identificado varias entidades que considera partes de una organización (ORG) en el texto. La presencia de prefijos como "##" indica que algunas de estas entidades pueden ser partes de palabras que siguen a la entidad principal.








**Generación de Resúmenes Abstractivos**

```python
from transformers import T5ForConditionalGeneration, T5Tokenizer

# Cargar el modelo y el tokenizador T5
model_name = "t5-small"
tokenizer = T5Tokenizer.from_pretrained(model_name)
model = T5ForConditionalGeneration.from_pretrained(model_name)

# Texto de entrada
text = """
El objetivo principal de este proyecto en Power BI es analizar y visualizar las principales tendencias de ventas de Adidas durante los años 2020 y 2021. A través de la creación de dashboards interactivos y gráficos intuitivos, se busca proporcionar una comprensión clara y detallada de la dinámica de ventas, identificando patrones, fluctuaciones estacionales y puntos destacados en la actividad comercial.
"""

# Tokenizar el texto de entrada
inputs = tokenizer.encode("summarize: " + text, return_tensors="pt", max_length=512, truncation=True)

# Generar el resumen
summary_ids = model.generate(inputs, max_length=150, min_length=40, length_penalty=2.0, num_beams=4, early_stopping=True)

# Decodificar y mostrar el resumen
summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
print("Resumen del texto:")
print(summary)
```

La salida del resumen es similar a la anterior, mejorando la realizada por primera vez.
